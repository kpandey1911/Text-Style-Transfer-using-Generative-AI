```python
import os
from getpass import getpass
import types
import pandas as pd
from botocore.client import Config
import ibm_boto3
from pandas import read_csv
from ibm_watson_machine_learning.foundation_models.utils.enums import ModelTypes
from ibm_watson_machine_learning.metanames import GenTextParamsMetaNames as GenParams
from ibm_watson_machine_learning.foundation_models import Model
```

```python
credentials = {
    "url" : "https://us-south.ml.cloud.ibm.com",
    "apikey" : getpass("Enter your API key: ")
}
```

```python
try:
    project_id = os.environ["PROJECT_ID"]
except KeyError:
    project_id = input("Enter your Project ID: ")
```

```python
project_id
```

```python
model_id = ModelTypes.FLAN_T5_XXL
```

```python
parameters={
    GenParams.DECODING_METHOD:"greedy",
    GenParams.RANDOM_SEED:33,
    GenParams.REPETITION_PENALTY:1,
    GenParams.MIN_NEW_TOKENS:5,
    GenParams.MAX_NEW_TOKENS:1024
}
```

```python
model = Model(
    model_id=model_id,
    params=parameters,
    credentials=credentials,
    project_id=project_id
)
```

```python

article = """
India is the world's largest democracy and one of the fastest-growing major economies. 
With a population exceeding 1.4 billion people, it is a hub of cultural, linguistic, and religious diversity. 
India's economy has seen rapid growth in sectors like IT, pharmaceuticals, and manufacturing. 
The government has also been focusing on infrastructure development and digital transformation initiatives to support sustainable growth.
"""


prompt = f"Summarize the following: {article}"

generated_response = model.generate(prompt=prompt)
```

```python
print("🔍 Summary:\n", generated_response['results'][0]['generated_text'])
```
